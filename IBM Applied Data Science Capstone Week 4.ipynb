## 1. Business Problem

The objective of the present project is to use Machine Learning algorithms to do a comparison of neighborhoods in Brussels. I will use descriptive statistics and non-supervised clustering algorithms in order to evaluate neighborhoods in Brussels according to specific criteria. The selection criteria are going to be based on what I look for in terms of availability of services when looking for accomodation in a city.

The selection criteria are going to be based on the availability of Supermarkets, Gym/Fitness centers, pharmacies, and Bars and Restaurants.





```python
!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim 


from IPython.display import Image 
from IPython.core.display import HTML 
    
!conda install -c conda-forge folium=0.5.0 --yes
import folium 

print('Folium installed')
print('Libraries imported.')
```

```python
address = 'Rue Montagne aux Herbes Potagères 26, 1000 Bruxelles, Belgium'

geolocator = Nominatim(user_agent="battle-of-the-neighborhoods")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude,longitude)
```

```python
bxl_map = folium.Map(location=[latitude, longitude], zoom_start=13)

folium.features.CircleMarker(
    [latitude, longitude],
    radius =10,
    color='red',
    popup='KU Leuven Campus Brussel',
    fill = True,
    fill_color ='red',
    fill_opacity = 0.6,
).add_to(bxl_map)

bxl_map
```

Above you can see the map of Brussels, with KU Leuven marked.
